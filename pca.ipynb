# PCA Analysis and Visualization on the Iris Dataset

#### This script performs Principal Component Analysis (PCA) on the Iris dataset, visualizes the explained variance of each principal component

the following libraries are required:

• matplotlib — for data visualization.

• seaborn — for making it easier to create graphs using matplotlib.

• scikit-learn — for working with machine learning methods, including
PCA and loading datasets.

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import seaborn as sns
```

 Data Loading

```python
data = load_iris()
X = data.data
y = data.target
```

We want to visually understand which of the components have a high variance.

Using methods:
- `pca_full.fit_transform(X)`: Performs PCA and extracts the main components.  
- `pca_full.explained_variance_ratio_`: Shows the proportion of variance explained by each component.

```python
pca_full = PCA()
X_pca_full = pca_full.fit_transform(X)
plt.figure(figsize=(8, 4))
plt.bar(range(1, len(pca_full.explained_variance_ratio_) + 1),
pca_full.explained_variance_ratio_)
plt.xlabel('Principal Components')
plt.ylabel('Variance Explained')
plt.title('Scree Plot')
plt.show()
```

The graph above shows that the first and second components make the main contribution.

Projection onto new coordinates

We project from 4-dimensional to 2-dimensional space and this time specifying the number of components in the line `pca = PCA(n_components=2)`.

`sns.scatterplot` is a method from the Seaborn library that creates a scatter plot.

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=y,
palette='viridis')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('2D PCA Scatter Plot')
plt.show()
```

In conclusion, we see that Setosa (0) is separated from the others, which means its features are significantly different. Versicolor (1) and Virginica (2) partially overlap, indicating that their features are similar.
